```python
from pathlib import Path

game_sales = Path("Video_Games.csv")
game_sales_df = pd.read_csv(game_sales)
game_sales_df.head()
```

```python
# Sorting the df to only include games with Global_sales >= 1 million
top_game_sales = game_sales_df[game_sales_df['Global_Sales'] >=1]
top_game_sales

# remove all nan values
top_game_sales_cleaned = top_game_sales.dropna()
top_game_sales_cleaned
```

```python
# Calculate total number of platforms
platform_count = top_game_sales_cleaned['Platform'].nunique()
platform_count
```

```python
# Calculate total number of publishers
publisher_count = top_game_sales_cleaned['Publisher'].nunique()
publisher_count
```

```python
# Developer Count
developer_count = top_game_sales_cleaned['Developer'].nunique()
developer_count
```

```python
# Total number of games 
game_count = top_game_sales_cleaned['Name'].nunique()
game_count
```

```python
# Calculate Total sales for each region
na_sales= sum(top_game_sales_cleaned['NA_Sales'])
eu_sales= sum(top_game_sales_cleaned['EU_Sales'])
jp_sales= sum(top_game_sales_cleaned['JP_Sales'])
other_sales = sum(top_game_sales_cleaned['Other_Sales'])
total_sales= na_sales + eu_sales + jp_sales + other_sales

total_sales_df = pd.DataFrame([
    {'Total of NA': na_sales, 
     'Total of EU': eu_sales, 
     'Total of JP': jp_sales, 
     'Total of Other': other_sales, 
     'Total Sales':total_sales}
])

dfi.export(total_sales_df, 'Espy_images/totals_tables.png')
total_sales_df
```

```python
# % NA from Total
percent_na = na_sales / total_sales
percent_na
```

```python
# average critic score
avg_critic_score = top_game_sales_cleaned['Critic_Score'].mean()
avg_critic_score
```

```python
# Calculate percentage of games with higher than 80% critic score
positive_critic_review_count = top_game_sales_cleaned[(top_game_sales_cleaned['Critic_Score'] >= 80)].count()["Name"]
positive_critic_game_percentage  = positive_critic_review_count / float(game_count) * 100
positive_critic_game_percentage
```

```python
# Calculate percentage of games with higher than 8 User score
positive_user_review_count = top_game_sales_cleaned[(top_game_sales_cleaned['User_Score'].astype('float64') >= 8)].count()["Name"]
positive_user_game_percentage  = positive_user_review_count / float(game_count) * 100
positive_user_game_percentage
```

```python
game_sales_summary = pd.DataFrame([
    {'Total Number of Platforms': platform_count, 
     'Total Number of Publishers': publisher_count, 
     'Total Number of Games': game_count, 
     'Total Sale From NA': na_sales, 
     'Total Sales From EU': eu_sales, 
     'Total Number of Sales From JP': jp_sales, 
     'Total Number of Sales from Other regions': other_sales, 
     'Total Sales': total_sales, 
     'Over 80% Critic Score' : positive_critic_game_percentage, 
     'Over 80% User Score': positive_user_game_percentage}
])

dfi.export(game_sales_summary, 'Espy_images/data_summary_table.png')
game_sales_summary
```

```python
# pie chart for each region's % of total sales
plt.figure(figsize=(8, 8))
plt.pie(sales_df['Sales'], labels=sales_df['Region'], autopct='%1.1f%%', startangle=140)
plt.title('% of each region to the Total')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.savefig('Espy_images/region%')
plt.show()
```

```python
# Create separate pie charts for each region to visualize the distribution of sales among the top 10 games 
# in that region

top_10_na = top_game_sales.nlargest(10, 'NA_Sales')
top_10_eu = top_game_sales.nlargest(10, 'EU_Sales')
top_10_jp = top_game_sales.nlargest(10, 'JP_Sales')
top_10_other = top_game_sales.nlargest(10, 'Other_Sales')
top_10_global = top_game_sales.nlargest(10, 'Global_Sales')

# Function to create a pie chart for a given region
def create_pie_chart(region_df, region_name):
    plt.figure(figsize=(8, 8))
    plt.pie(region_df['Global_Sales'], labels=region_df['Name'], autopct='%1.1f%%', startangle=140)
    plt.title(f'Top 10 Games in {region_name}')
    plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
    plt.savefig('Espy_images/Top_na')
    plt.show()
    

# Call the function for each region
create_pie_chart(top_10_na, 'North America')

```
